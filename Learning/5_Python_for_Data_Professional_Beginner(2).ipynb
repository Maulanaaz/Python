```python
#List Manipulations

#Merging list
prices = [6000,7500,3000,2300]
items = ["Chicken Noodle","Meatball","Hot Rice","Ice Tea"]
menu = items + prices
print(menu)

#List Methods
'''''
Python has a set of built-in methods that you can use on lists.

Method 	 Description
append() 	 Adds an element at the end of the list
clear() 	 Removes all the elements from the list
copy() 	 Returns a copy of the list
count() 	 Returns the number of elements with the specified value
extend() 	 Add the elements of a list (or any iterable), to the end of the current list
index() 	 Returns the index of the first element with the specified value
insert() 	 Adds an element at the specified position
pop() 	 Removes the element at the specified position
remove() 	 Removes the item with the specified value
reverse() 	 Reverses the order of the list
sort() 	 Sorts the list
'''''
languages = ['Python','Java','C++']
print("Languages = ",languages)
languages.append('C') #APPEND
print("Appended Languages = ",languages)
languages.clear() #CLEAR
print("Cleared Languages = ",languages)
languages = ['Python','Java','C++']
languages2 = languages.copy() #COPY
print("Copied Languages = ",languages2)
languages = ['Python','Java','C++','Python','Java','Python','C','Ruby']
print("Updated Languages = ",languages)
print("Python is listed ",languages.count('Python')," Times") #COUNT
new_language = ['Kotlin']
print("New Language = ",new_language)
languages.extend(new_language) #EXTEND
print("Extended languages = ",languages)
print("C++ is in index ",languages.index('C++')) #INDEX
languages.insert(3,"Lost") #INSERT
print("Updated languages = ",languages)
languages.pop(3) #POP
print("Updated languages = ",languages)
languages.remove('Kotlin') #REMOVE
print("Updated languages = ",languages)
languages.reverse() #REVERSE
print("Reversed languages = ",languages)
languages.reverse()
languages.sort() #ASCENDING SORT
print("Sorted languages (Ascending) = ",languages)
languages.sort(reverse=True) #DESCENDING SORT
print("Sorted languages (Descending) = ",languages)
```

```python
# Methods in Sets
'''''
Python has a set of built-in methods that you can use on sets.

Method 	 Description
add() 	 Adds an element to the set
clear() 	 Removes all the elements from the set
copy() 	 Returns a copy of the set
difference() 	 Returns a set containing the difference between two or more sets
difference_update() 	 Removes the items in this set that are also included in another, specified set
discard() 	 Remove the specified item
intersection() 	 Returns a set, that is the intersection of two other sets
intersection_update() 	 Removes the items in this set that are not present in other, specified set(s)
isdisjoint() 	 Returns whether two sets have a intersection or not
issubset() 	 Returns whether another set contains this set or not
issuperset() 	 Returns whether this set contains another set or not
pop()	 Removes an element from the set
remove() 	 Removes the specified element
symmetric_difference() 	 Returns a set with the symmetric differences of two sets
symmetric_difference_update() 	 inserts the symmetric differences from this set and another
union() 	 Return a set containing the union of sets
update() 	 Update the set with the union of this set and others
'''''
fruit = {"Apple","Orange","Lemon","Mango"}
print("Set of fruits = ",fruit)
fruit.add("Watermelon") #ADD
print("New set of fruits = ",fruit)
fruit2 = fruit.copy() #COPY
fruit3 = fruit
print("Is fruit 3 equal to fruit ?", fruit == fruit3)
print("Copied fruit = ",fruit2)
print("Is fruit 2 equal to fruit ?", fruit == fruit2)
fruit.pop() #POP
print("Popped set of fruits : ",fruit)
vegetable = {"Cabbage","Cucumber","Carrot"}
fruit.update(vegetable) #UPDATE
print("Updated set of fruits & vegetables = ",fruit)
fruit.remove("Cucumber") #REMOVE
print("New set of fruits & vegetables = ",fruit)

fruit = {"Apple","Orange"}
fruit2 = {"Lemon","Mango"}
fruit3 = {"Apple","Avocado","Orange","Lemon"}
fruit4 = fruit.union(fruit2) #UNION
print("fruit 4 = ",fruit4)
## ISDISJOINT ##
print(fruit.isdisjoint(fruit2)) #Will return True because of there is no same value between fruit and fruit2
print(fruit.isdisjoint(fruit3)) #Will return False because of 'fruit' has same value in 'fruit3' (Apple and Orange)

## ISSUBSET ##
print(fruit.issubset(fruit3)) #Will return True because of all of the 'fruit' values are in 'fruit3'
print(fruit.issubset(fruit2)) #Will return False because of all of the 'fruit' values are not in 'fruit2'

## ISSUPERSET ##
print(fruit3.issuperset(fruit)) #Will return True because of all of the 'fruit' values are in 'fruit3'
print(fruit3.issuperset(fruit2)) #Will return False because of all of the 'fruit' values are not in 'fruit2'

## DIFFERENCE ##
print(fruit3.difference(fruit)) #Will return 'fruit3's values that arent exist in 'fruit' 
print(fruit.difference(fruit2)) #Will return 'fruit3's values that arent exist in 'fruit2' 

## INTERSECTION ##
print(fruit3.intersection(fruit)) #Will return intersection value(s) between 'fruit3' and 'fruit'
print(fruit3.intersection(fruit2)) #Will return intersection value(s) between 'fruit3' and 'fruit'

## SYMMETRIC_DIFFERENCE
fruit = {"Apple","Orange","Melon","Tomato"}
fruit2 = {"Lemon","Mango","Cherry","Lychee"}
fruit3 = {"Apple","Avocado","Orange","Lemon"}
print(fruit3.symmetric_difference(fruit)) #Will return union of 'fruit' and 'fruit3's difference
print(fruit3.symmetric_difference(fruit2)) #Will return union of 'fruit2' and 'fruit3's difference
```

```python
# Methods in Dictionaries
'''''
Python has a set of built-in methods that you can use on dictionaries.

Method 	 Description
clear() 	 Removes all the elements from the dictionary
copy() 	 Returns a copy of the dictionary
fromkeys() 	 Returns a dictionary with the specified keys and value
get() 	 Returns the value of the specified key
items() 	 Returns a list containing a tuple for each key value pair
keys() 	 Returns a list containing the dictionary's keys
pop() 	 Removes the element with the specified key
popitem() 	 Removes the last inserted key-value pair
setdefault() 	 Returns the value of the specified key. If the key does not exist: insert the key, with the specified value
update() 	 Updates the dictionary with the specified key-value pairs
values() 	 Returns a list of all the values in the dictionary
'''''
data1 = {
    'Name' : 'Zulfikar',
    'Age' : 16,
    'Country' : 'Indonesia'
}
data1.clear() #CLEAR
print("Cleared data1 = ",data1)
data2 = {
    'Name' : 'Maulana',
    'Age' : 16,
    'Country' : 'Indonesia'
}
data3 = data2.copy() #COPY
print("Data 3 copied from data 2 = ",data3)
print(list(data2.keys())) #KEYS = Return all keys that exist in the dictionary
print(list(data2.values())) #VALUES = Return all values that exist in the dictionary
data2.update({"Language" : "Python"}) #UPDATE
print("Updated data 2 = ",data2)
```

```python
# File Handling
'''''
The key function for working with files in Python is the open() function.
The open() function takes two parameters; filename, and mode.
There are four different methods (modes) for opening a file:
"r" - Read - Default value. Opens a file for reading, error if the file does not exist
"a" - Append - Opens a file for appending, creates the file if it does not exist
"w" - Write - Opens a file for writing, creates the file if it does not exist
"x" - Create - Creates the specified file, returns an error if the file exist

In addition you can specify if the file should be handled as binary or text mode
"t" - Text - Default value. Text mode
"b" - Binary - Binary mode (e.g. images)
'''''
print("-----------")
print("Read")
print("-----------")
file = open("Document/example.txt","r")
content = file.read()
file.close() #Close the document
print(content)

#You can return one line by using the readline() method
file = open("Document/example.txt","r")
first_content = file.readline()
second_content = file.readline()
file.close()
print(first_content)
print(second_content)

#By using readlines() method you can return a list containing each line in the file as a list item
file = open("Document/example.txt","r")
all_files = file.readlines()
file.close()
print(all_files)

#You can use for loop to read file
file = open("Document/example.txt","r")
for line in file :
    print(line)
file.close()
print("-----------")
print("Write")
print("-----------")

newfile = open("Document/example_2.txt","w") #Opens a file for writing, creates the file if it does not exist
newfile.write("This is how we write txt file by Python")
newfile.close()
file = open("Document/example_2.txt","r")
content = file.read()
file.close()
print(content)

print("-----------")
print("Append")
print("-----------")
append_file = open("Document/example_2.txt","a") #Opens a file for appending, creates the file if it does not exist
append_file.writelines([
    "\nAnd this is how we append txt file by Python \n",
    "What a powerful function!"
]) #Opens a file for appending, creates the file if it does not exist
append_file.close()
file = open("Document/example_2.txt","r")
content = file.read()
file.close()
print(content)
```
