```python
# Membaca Data Survey
import numpy as np
data = np.loadtxt("https://storage.googleapis.com/dqlab-dataset/survei_tinggi_badan.txt",skiprows=1)
print("Ukuran Data = {}".format(data.shape))
print(data)
```

```python
# Ukuran Variasi : Range
print("min data :",data.min())
print("max data :",data.max())
```

```python
# Visualisasi bulir data
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(9,3))
sns.swarmplot(x=data,size=7,color="grey",ax=ax,edgecolor=None)
ax.grid(axis="x")
ax.set_xlabel("Tinggi Badan (CM)")
plt.tight_layout()
plt.show()
```

```python
# Mencari Nilai Quartil dengan Menggunakan np.percentile()
Q1 = np.percentile(data,25)
print("Quartil Pertama dari data survey tersebut adalah :",Q1)
Q2 = np.percentile(data,50)
print("Quartil Kedua dari Data Survey Tersebut Adalah :",Q2)
Q3 = np.percentile(data,75)
print('Quartil Ketiga dari Data Survey Tersebut Adalah :',Q3)

Q123 = np.percentile(data,[25,50,75])
print("Quartil Pertama, Kedua, dan Ketiga dari Data Survey Tersebut Adalah : {}".format(Q123))
```

```python
# Mencari Nilai Quartil dengan Menggunakan np.quantile()
Q1 = np.quantile(data,0.25)
Q2 = np.quantile(data,0.50)
Q3 = np.quantile(data,0.75)
print("Quartil Pertama dari data survey tersebut adalah :",Q1)
print("Quartil Kedua dari data survey tersebut adalah :",Q2)
print("Quartil Ketiga dari data survey tersebut adalah :",Q3)
Q123 = np.quantile(data,[0.25,0.50,0.75])
print("Quartil Pertama, Kedua, dan Ketiga dari data tersebut adalah :",Q123)
```

```python
# Mencari Nilai Upper Whisker dan Lower Whisker
IQR = Q3 - Q1
lower_whisker = Q1 - 1.5*IQR
upper_whisker = Q3 + 1.5*IQR
print("Lower Whisker dari data survey tersebut adalah :",lower_whisker)
print("Upper Whisker dari data survey tersebut adalah :", upper_whisker)
```

```python
# Visualisasi Boxplot
fig, ax = plt.subplots(figsize=(10,4))
sns.swarmplot(x=data,size=7,color="darkcyan",edgecolor=None,ax=ax)
sns.boxplot(x=data,whis=1.5,color="greenyellow",ax=ax)
```

```python
# Mencari Persentil ke-10 dan ke-70 dengan menggunakan np.percentile() dan np.quantile()
persentil_10_1 = np.percentile(data,10)
persentil_10_2 = np.quantile(data,0.1)
persentil_70_1 = np.percentile(data,70)
persentil_70_2 = np.percentile(data,70)

if persentil_10_1 == persentil_10_2 :
    print("Persentil ke-10 dari data tersebut adalah: ", persentil_10_1)
if persentil_70_1 == persentil_70_2 :
    print("Persentil ke-70 dari data tersebut adalah: ", persentil_70_1)
```

```python
# Mencari Standar Deviasi dan Varians

def rata_rata(data) :
    n = len(data)
    return sum(data) / n

def varians(data,k=1) :
    rerata, var = rata_rata(data), 0
    for i in data :
        var += (i - rerata)**2
    return var / (len(data)-k)

def standar_deviasi(data) :
    return varians(data) ** 0.5

varians_1 = varians(data) 
varians_2 = data.var()
std_1 = standar_deviasi(data)
std_2 = data.std()

print("Varians dengan menggunakan cara manual : ", varians_1) # Unbiased
print("Varians dengan menggunakan np.var() : ", varians_2) # Biased
print("Standar Deviasi dengan menggunakan cara manual : ", std_1) # Unbiased
print("Standar Deviasi dengan menggunakan np.std() : ", std_2) # Biased
```

```python
# Problem 1 : Eksplorasi data dan mencari IQR serta persentil dengan menggunakan library pandas
import pandas as pd
data = pd.read_csv('https://storage.googleapis.com/dqlab-dataset/survei_tinggi_badan.txt')
print("Data Survey : \n")
print(data)
stat_desc = data.describe()
print("Statistik Deskriptif Data Survey :")
print(stat_desc)
IQR = stat_desc["tinggi badan (cm)"]["75%"] - stat_desc["tinggi badan (cm)"]["25%"]
print("Diperoleh IQR = ",IQR)
persentil_80 = data.quantile(q=0.8)
print("Persentil ke-80 dari data tersebut adalah : ", persentil_80)
```

```python
# Problem 2 : Membandingkan grafik boxplot sebelum dan sesudah modifikasi\
# Import pandas sebagai aliasnya yaitu
import pandas as pd

# Baca dataset
tinggi_badan = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/survei_tinggi_badan.txt")
print("Statistik deskriptif data awal:\n", tinggi_badan.describe().T)

# Modifikasi data maksimum tinggi badan menjadi 186 cm
tinggi_badan_mod = tinggi_badan.replace(tinggi_badan.max(), 186)
print("\nStatistik deskriptif data modifikasi:\n", tinggi_badan_mod.describe().T)

# Import matplotlib.pyplot dan seaborn 
import matplotlib.pyplot as plt
import seaborn as sns
fig, axs = plt.subplots(2, 1, figsize=(12,8), sharex=True)
# Plot data awal
sns.boxplot(data=tinggi_badan, x="tinggi badan (cm)", whis=1.5, color="yellowgreen", ax=axs[0])
sns.swarmplot(data=tinggi_badan, x="tinggi badan (cm)", size=8, color="darkcyan", edgecolor=None, ax=axs[0])
axs[0].grid(axis="x")
axs[0].set_xlabel(None)
axs[0].set_title("Data awal", color="darkcyan", fontsize=16)

# Plot data modifikasi
sns.boxplot(data=tinggi_badan_mod, x="tinggi badan (cm)", whis=1.5, color="yellowgreen", ax=axs[1])
sns.swarmplot(data=tinggi_badan_mod, x="tinggi badan (cm)", size=8, color="darkcyan", edgecolor=None, ax=axs[1])
axs[1].grid(axis="x")
axs[1].set_xlabel(None)
axs[1].set_title("Data modifikasi", color="darkcyan", fontsize=16)

plt.xlabel("Tinggi badan (cm)", fontsize=14)
plt.tight_layout()
plt.show()
```
