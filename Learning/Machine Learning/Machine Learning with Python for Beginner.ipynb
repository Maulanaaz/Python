```python
# Scaling 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaling_column = ['Administrative', 'Administrative_Duration', 'Informational', 'Informational_Duration', 'ProductRelated', 
                  'ProductRelated_Duration', 'BounceRates', 'ExitRates', 'PageValues']
data[scaling_column] = scaler.fit_transform(data[scaling_column])
print(data[scaling_column].describe().T[["min","max"]])
```

```python
# Encode feature yang memiliki tipe str
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Feature : Month
LE = LabelEncoder()
data["Month"] = LE.fit_transform(data["Month"])
print(LE.classes_)
print(np.sort(data["Month"].unique()))
print("")
# Feature : VisitorType
LE = LabelEncoder()
data["VisitorType"] = LE.fit_transform(data["VisitorType"])
print(LE.classes_)
print(np.sort(data["VisitorType"].unique()))
```

# Modelling

```python
# Memilih label dan features
X = data.drop("Revenue",axis=1)
y = data["Revenue"]
print("Ukuran X = {} dan Ukuran y = {}".format(X.shape,y.shape))
```

```python
# Menentukan training dataset dan testing dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
print("ukuran X train :",X_train.shape)
print("ukuran X test :",X_test.shape)
print("ukuran y train :",y_train.shape)
print("ukuran y test :",y_test.shape)
```

```python
# Menentukan model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
# Memprediksi testing dataset
y_pred = model.predict(X_test)
print(y_pred.shape)
```

```python
# Evaluasi Model
from sklearn.metrics import confusion_matrix, classification_report
# Evaluasi dari accuracy score
print("Nilai akurasi dari training dataset :", model.score(X_train, y_train))
print("Nilai akurasi dari testing dataset :", model.score(X_test, y_test))

# Evaluasi menggunakan confusion matrix
print("")
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix :")
print(cm)

# Menampilkan classification report
print("Classification report :")
cr = classification_report(y_test, y_pred)
print(cr)
```

# Pemodelan : Algoritma Logistic Regression

```python
import pandas as pd
import numpy as np
data = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/pythonTutorial/online_raw.csv")

# Drop NA Values
data = data.dropna()

# Rescaling
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
scale = ['Administrative', 'Administrative_Duration', 'Informational',
       'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration',
       'BounceRates', 'ExitRates', 'PageValues']
Scaler = MinMaxScaler()
data[scale] = Scaler.fit_transform(data[scale])
LE = LabelEncoder()
data["Month"] = LE.fit_transform(data["Month"])
print(LE.classes_)
print(np.sort(data["Month"].unique()))
LE = LabelEncoder()
data["VisitorType"] = LE.fit_transform(data["VisitorType"])
print(LE.classes_)
print(np.sort(data["VisitorType"].unique()))

# Modelling
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X = data.drop("Revenue",axis=1)
y = data["Revenue"]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

# Evaluasi Model
from sklearn.metrics import confusion_matrix, classification_report
print("Akurasi Model :",model.score(X_test,y_test))
cm = confusion_matrix(y_test,y_pred)
cr = classification_report(y_test,y_pred)
print("Confusion Matrix :")
print(cm)
print("Classification Report :")
print(cr)
```

# Pemodelan : Decision Tree Classifier

```python
import pandas as pd
import numpy as np
data = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/pythonTutorial/online_raw.csv")

# Mencari tahu apakah ada missing values
print("Total Missing Values Sebelum Impute:",data.isna().sum().sum())
# Mengimpute missing values dengan nilai rata-ratanya
data.fillna(data.mean(),inplace=True)
print("Total Missing Values Setelah Impute:",data.isna().sum().sum())
# Rescaling data continue
from sklearn.preprocessing import MinMaxScaler
rescale = ['Administrative', 'Administrative_Duration', 'Informational',
       'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration',
       'BounceRates', 'ExitRates', 'PageValues']
scaler = MinMaxScaler()
data[rescale] = scaler.fit_transform(data[rescale])
# Label Encoding data kategorikal
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
data["Month"] = LE.fit_transform(data["Month"])
print("Month :",LE.classes_)
print("Label :",np.sort(data["Month"].unique()))
LE = LabelEncoder()
data["VisitorType"] = LE.fit_transform(data["VisitorType"])
print("VisitorType :",LE.classes_)
print("Label :",np.sort(data["VisitorType"].unique()))
# Modelling
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X = data.drop("Revenue",axis=1)
y = data["Revenue"]
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,test_size=0.3)
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
# Predicting
y_pred = model.predict(X_test)
# Evaluasi Model
from sklearn.metrics import confusion_matrix, classification_report
print("Akurasi dari model :",model.score(X_test,y_test))
print("Confusion Matrix :")
cm = confusion_matrix(y_test,y_pred)
cr = classification_report(y_test,y_pred)
print(cm)
print("Classification Report :")
print(cr)
```

# Pemodelan : Linear Regression

```python
import pandas as pd
data = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/pythonTutorial/housing_boston.csv")
print(data.shape)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[['RM','LSTAT','PTRATIO','MEDV']] = scaler.fit_transform(data)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X = data.drop("MEDV",axis=1)
y = data["MEDV"]
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,test_size=0.2)
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

# Evaluasi
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("Akurasi dari Model :",model.score(X_test,y_test))
print("MSE dari model :",mean_squared_error(y_test,y_pred))
print("RMSE dari model :",mean_squared_error(y_test,y_pred,squared=False))
print("MAE dari model :",mean_absolute_error(y_test,y_pred))
print("R^2 dari model :",r2_score(y_test,y_pred))

# Plotting 
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred,c="red")
plt.title("Scatter Plot ")
```

# Pemodelan : K-Nearest Neighbor

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/pythonTutorial/mall_customers.csv")
# Rescaling 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[['annual_income', 'spending_score']] = scaler.fit_transform(data[['annual_income', 'spending_score']])
data_model = data[["annual_income","spending_score"]]
# Cari Nilai K yang optimal
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
k = 2
inertia_values = []
while k <= 10 :
    model = KMeans(n_clusters=k,random_state=0)
    clusters = model.fit_predict(data_model)
    model.fit(data_model)
    inertia_values.append(model.inertia_)
    print("Silhouette Score {}:".format(k),silhouette_score(data_model,clusters))
    k += 1
import matplotlib.pyplot as plt
plt.figure()
plt.plot(range(2,11),inertia_values,marker="o")
plt.title("Grafik Elbow Method")
plt.xlabel("Jumlah K")
plt.ylabel("Inersia")
plt.tight_layout()
plt.show()
# Modelling
model = KMeans(n_clusters=5,random_state=0)
label = model.fit_predict(data_model)
# Plotting
import numpy as np
X = np.array(data_model)
xs = X[:,0]
xy = X[:,1]
fig, ax = plt.subplots(1,2,figsize=(10,4))
ax[1].scatter(xs,xy,c=label,alpha=0.5) # Plot titik-titik data
centroids = model.cluster_centers_
xc = centroids[:,0]
yc = centroids[:,1]
ax[1].scatter(xc,yc,marker="D",s=50,c='red')
ax[1].set_title("Scatter Plot Data Hasil KNN")
ax[1].set_xlabel("Annual Income")
ax[1].set_ylabel("Spending Score")
ax[0].scatter(X[:,0],X[:,1])
ax[0].set_title("Scatter Plot Data")
ax[0].set_xlabel("Annual Income")
ax[0].set_ylabel("Spending Score")
plt.tight_layout()
plt.show()
```
