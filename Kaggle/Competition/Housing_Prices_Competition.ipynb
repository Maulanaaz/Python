# Importing the needed libraries

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

# Calling our data (s)

```python
train_data = pd.read_csv('dataset/train.csv')
test_data = pd.read_csv('dataset/test.csv')
```

# Exploring our data

```python
# train_data
print(train_data.columns)
train_data.describe()
```

```python
# test_data
print(test_data.columns)
test_data.describe()
```

# Let's choose the target and features !

```python
"""
The purpose here is to predict the final home's price, so we use 'SalePrice' column for our target
From the given columns description, we can choose these columns to be our features : 'LotArea', 'OverallQual',
'OverallCond', 'YearBuilt', 'YearRemodAdd', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'FullBath', 'TotRmsAbvGrd'
"""

y = train_data.SalePrice
features = ['LotArea','OverallQual','OverallCond','YearBuilt','YearRemodAdd',
           '1stFlrSF','2ndFlrSF','GrLivArea','FullBath','TotRmsAbvGrd']
X = train_data[features]
train_X, val_X, train_y, val_y = train_test_split(X,y,random_state=1) # splitting our train_data
```

# Let's build our model !

```python
model = RandomForestRegressor(random_state=1)
model.fit(train_X,train_y)
```

# Let's validate our model !

```python
mean_absolute_error(val_y,model.predict(val_X))
```

It seems acceptable..

# It's time to make the prediction !

```python
test_data2 = test_data[features] #filtering the test_data based on the features we have chosen
prediction = model.predict(test_data2)
```

# Finally, let's create a file to submit our prediction to the competition !

```python
output = pd.DataFrame({'Id':test_data.Id,
                      'SalePrice': prediction})
output.to_csv('maulana_submission.csv',index=False)
```
