```python
# operator overloading is when a desginer makes its own operation use operators (+ - * / etc)
import numpy as np
a = np.array([1,2,3,4,5,6])
b = [1,2,3,4,5,6]
print(a + 2) # it will make each element in the array operated by "+" because the designer designed it
print(b+2) # will raise error because the designer did not allow list to be "+" operated with integer 
```

```python
def yoyoi (hand_1,hand_2) :
    # point for hand_1
    point_1 = 0
    point_2 = 0
    total_a = 0
    total_a2 = 0
    for i in hand_1 :
        if i in ["K","Q","J"] :
            point_1 += 10
        elif i == "A" :
            point_1 += 11
            total_a += 1
        else :
            point_1 += int(i)
    if total_a != 0 and point_1 > 21 :
        while point_1 > 21 :
            point_1 -= 10
            total_a -= 1
            if total_a == 0 and point_1 > 21 :
                return False
    # point for hand 2
    for i in hand_2 :
        if i in ["K","Q","J"] :
            point_2 += 10
        elif i == "A" :
            point_2 += 11
            total_a2 += 1
        else :
            point_2 += int(i)
    if total_a2 != 0 and point_2 > 21 :
        while point_2 > 21 :
            point_2 -= 10
            total_a2 -= 1
            if total_a2 == 0 and point_2 > 21 :
                print("Here")
                return True
    print("Yes")
    if total_a > 21 :
        return False
    elif total_a2 > 21 :
        return True
    print("No")
    return point_1 > point_2

print(yoyoi(['J', '2', '9', '5'],['A', '8', '8', 'Q', '4', '7']))
```
