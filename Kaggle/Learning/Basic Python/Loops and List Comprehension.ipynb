```python
# Loops are the way to repeatedly execute some code

# list
fruit = ["Banana","Dragonfruit","Durian","Pineapple","Mango","Lemon"]
for i in fruit :
    print(i, end=" ")

# Tuple
grades = (89,95,79,94,56,87,82,80,91,90)
print("\n")
for i in grades :
    print(i, end="-")
    
# String
attention = "Please choose your way of life, the way that can make you better. do not forget who you are, or you will regret"
print("\n")
for word in attention :
    print(word, end=" ")
```

```python
# Using range() function on for loop

for i in range(10) :
    print(i,end=" ")
print("\n")
for i in range(0,10) :
    print(i, end=" ")
print("\n")
for i in range(0,11,2) :
    print(i, end=" ")
```

```python
# Another type of loop in python is while loop, where it will iterate until the condition met
aldo_grade = 70
remidy = 0
while aldo_grade < 85 :
    aldo_grade += 2
    remidy += 1
    print(str(aldo_grade) + " remidy")
```

```python
# List comprehension vs regular code
storage_amount = [50,43,56,78,32,43,10]
# Regular Code
storage_amounts = [50,43,56,78,32,43,10]
for i in range(len(storage_amounts)) :
    storage_amounts[i] -= 5
print(storage_amounts)
# list comprehension
storage_amounts = [i-5 for i in storage_amount]
print(storage_amounts)
```

```python
# List comprehension with if condition
grades = [50,40,20,70,89,40,95,38,78,69,93,29,91,94,82,48,85,86,38,80,98,87,86]
passed_grades = [i for i in grades if i >= 85]
print(passed_grades)

# List comprehension with if else condition
grades = [50,40,20,70,89,40,95,38,78,69,93,29,91,94,82,48,85,86,38,80,98,87,86]
explanation = ["Passed" if i >= 85 else "Not Passed" for i in grades]
print(explanation)
```

```python
# Transform list using list comprehension
grades = [50,40,20,70,89,40,95,38,78,69,93,29,91,94,82,48,85,86,38,80,98,87,86]
desc = [str(i)+" is enough" if i >= 85 else str(i) + " is not enough" for i in grades]
print(desc)
```

```python
# Minimalizing the length of our code

numbers = [9,12,3,-2,-9,0,9,-3,-4]

# The goal is to find how many the negative numbers in the list

# Traditional method
def traditional_method(num) :
    total = 0
    for i in num :
        if i < 0 :
            total += 1
    return total

# list comprehension
def comprehension(num) :
    return sum([i < 0 for i in num])

print(traditional_method(numbers))
print(comprehension(numbers))
```
