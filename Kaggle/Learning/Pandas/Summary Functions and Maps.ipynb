# Summary Your Data

```python
# Describing your data
import pandas as pd
data = pd.read_csv("Dataset/Data_countries.csv")
data.describe()
```

```python
# Describing your numerical data
print("Here is the forecast of the world population in 2050")
data.pop2050.describe()
```

```python
# Describing your text data
print("Hm... what is our country coloumn looks like?")
data.country.describe()
```

```python
# Other specific functions
print("The mean of population in 2022 is {}".format(data.pop2022.mean()))
print("There is the unique country we got {}".format(data.country.unique()))
print("Total of the country is {} countries".format(data.country.count()))
```

# Map

```python
# Using map() to transform our data

# Let's say we want to make a new column called difference_2022_2023 with its value is population in 2022 minus population in 2023
data["difference_2022_2023"] = data.pop2022.map(lambda p : p - data.pop2023 )
data.difference_2022_2023
```

```python
# Using apply() to transform our data
def pop_2050_1980(data) :
    data.pop_2050_1980 = data.pop2050 - data.pop1980
    return data

# we will make a column again named pop_2050_1980 which indicated how much the difference between population in 2050 with 1980
data["pop_2050_1980"] = pd.Series([0 for i in range(len(data))])
data.apply(pop_2050_1980, axis="columns")
```

# Built-ins Mapping Operations

```python
# Adding numerical data with another numerical data
data["pop2010_plus_pop2000"] = data.pop2010 + data.pop2000
data.pop2010_plus_pop2000
```
