# Getting Started

```python
import pandas as pd
```

# Creating Data (DataFrame)

```python
# Dataframe is a table contains rows and columns with values in it
statistics_score = pd.DataFrame({'Zhao':[96,94,92,93,94],'Dikky':[92,94,90,86,90]})
print(statistics_score)
statistics_score = pd.DataFrame({'Zhao':[96,94,92,93,94],'Dikky':[92,94,90,86,90]},index=["Hypothesis","Calculus","Programming and Algorithm","Descriptive Statistics","Sampling"])
print(statistics_score)
```

# Series

```python
# Series is a list. If a DataFrame is a table, series is a list in the table, it means that DataFrame is combined series(s)
mathematics_zhao = pd.Series([89,87,90,85,83])
print(mathematics_zhao)
mathematics_zhao = pd.Series([89,87,90,85,83],index=["Calculus I","Calculus II","Calculus III","Algebra I","Algebra II"],name = "Zhao's Mathematics Score")
print(mathematics_zhao)
```
