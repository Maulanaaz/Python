```python
# What is machine learning?
"""
machine learning is when machine use past data(s) to predict the new data(s)
one of examples of machine learning is decision tree
decision tree works like roots
it will predict the new data(s) using past data(s) by searching the pattern on the past data(s)
"""
```

```python
# Using pandas to get familiar with your data
import pandas as pd
df = pd.read_csv("dataset/melb_data.csv")
df
```

```python
# Interprete your data
df.describe()
```

```python
# Selecting data for modelling
df.dropna(axis=0) # drop the NA tables
df.columns #reveals all of our data's columns
```

```python
# Selecting prediction target
# prediction target is the variable that we want to predict, here we choose the price column
y = df.Price
```

```python
y.describe()
```

```python
# Selecting features
# features are the variables we want to use to predict the target variable
features = ['Rooms','Bathroom','Landsize','Lattitude','Longtitude']
X = df[features]
```

```python
X.describe()
```

```python
# Building our model using scikit-learn
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=1)
model.fit(X,y)
```

```python
# Let's predict the 5 first data on our dataframe !
print(X.head())
print(y.head())
print(model.predict(X.head()))
```
