# 1. Buatlah visualisasi 2-dimensi atau 3-dimensi dari vector berikut
- a. ùëù = (6, 3) dan ùëû = (3, 5)
- b. ùë• = (8, 13, 1), ùë¶ = (13, 1, 8), dan ùëß = (1, 8, 13)

```python
# a. p = (6,3) dan q = (3,5)

# Definisikan vector
p = np.array([[6],
             [3]])
q = np.array([[3],
             [5]])
plt.plot([0,p[0]],[0,p[1]],'b',label="vektor p") # Plotting vektor p
plt.plot([0,q[0]],[0,q[1]],'r',label="vektor q") # Plotting vektor q
plt.axis("square")
plt.axis([-2,7,-2,7])
plt.legend()
plt.grid()
plt.show()
```

```python
# b. ùë• = (8, 13, 1), ùë¶ = (13, 1, 8), dan ùëß = (1, 8, 13)
# Definisikan vektor
x = np.array([[8],
             [13],
             [1]])
y = np.array([[13],
             [1],
             [8]])
z = np.array([[1],
             [8],
             [13]])

fig = plt.figure(figsize = plt.figaspect(1))
ax = fig.add_subplot(111,projection='3d')

#Plotting vektor
ax.plot([0,x[0]],[0,x[1]],[0,x[2]],'r',label="vektor x",linewidth=3)
ax.plot([0,y[0]],[0,y[1]],[0,y[2]],'g',label="vektor y",linewidth=3)
ax.plot([0,z[0]],[0,z[1]],[0,z[2]],'b',label="vektor z",linewidth=3)

# plotting sumbu imajiner
ax.plot([-15,15],[0,0],[0,0],"--",color=[.6,.6,.6]) # sumbu x
ax.plot([0,0],[-15,15],[0,0],"--",color=[.6,.6,.6]) # sumbu y
ax.plot([0,0],[0,0],[-15,15],"--",color=[.6,.6,.6]) # sumbu z

ax.plot
ax.legend()
ax.set_xlim3d(-15,15)
ax.set_ylim3d(-15,15)
ax.set_zlim3d(-15,15)
plt.show()
```

# 2. Berdasarkan vector-vector di atas, tentukan

```python
# a. ùëù ‚ãÖ (8ùëù + ùëû) ‚àí (4ùë• + 7ùë¶)
print("Hasil dari  ùëù ‚ãÖ (8ùëù + ùëû) ‚àí (4ùë• + 7ùë¶) adalah :")
display(Math(sym.latex(sym.simplify(np.dot(p.T,(8*p)+q)-((4*x)+(7*y))))))
```

```python
# b. x ‚ãÖ y ‚ãÖ z

print("Hasil dari x ‚ãÖ y ‚ãÖ z adalah :")
display(Math(sym.latex(sym.simplify(np.dot(x.T,y,z)))))
```

# 3. Diketahui vektor ùëé = (1,2) dan ùëè = (4, 2) dan ùúÉ adalah sudut antara vektor ùëé dan ùëè. Tentukan nilai cos ùúÉ, sudut dalam radian, dan sudut dalam derajat

```python
# Definisikan vektor
a = np.array([[1],
             [2]])
b = np.array([[4],
             [2]])

# rumus sudut antara dua vektor
cos_teta = np.dot(a.T,b)/ (np.linalg.norm(a)* np.linalg.norm(b))
radians = mt.acos(cos_teta) # mencari teta dalam bentuk radian
degrees = radians * (180/mt.pi) # mengubah radian ke bentuk derajat
print("Nilai cos ùúÉ = ",round(cos_teta[0][0],2))
print("Sudut dalam radian = ",round(radians,3))
print("Sudut dalam derajat = {}¬∞ ".format(round(degrees,3)))
```

# 4. Diketahui vektor ùëé = (6, 11, 6), ùëè = (2, 1, 4), dan ùëê = (1, ‚àí1, 3). Berdasarkan vektor-vektor tersebut, hitunglah norm Euclidean dari masing-masing vector

```python
# Definisikan Vektor
a = np.array([[6],
             [11],
             [6]])
b = np.array([[2],
             [1],
             [4]])
c = np.array([[1],
             [-1],
             [-3]])
print("Norm Euclidean dari vektor a = (6,11,6) adalah ",round(np.linalg.norm(a),3))
print("Norm Euclidean dari vektor b = (2,1,4) adalah ",round(np.linalg.norm(b),3))
print("Norm Euclidean dari vektor c = (1,-1,3) adalah ",round(np.linalg.norm(c),3))
```
